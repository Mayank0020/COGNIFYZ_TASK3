## Develop a machine learning model to classify restaurants based on their cuisines.

## 1. Data Preprocessing

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB
```

```python
df2 = pd.read_csv(r"Dataset .csv")
```

```python
df2
```

```python
df2.info()
```

```python
df2.isna().sum()
```

```python
import re
def remove_special(text):
    if isinstance(text, str):
        return re.sub(r'[^\w\s]','', text)
    else:
        return text
```

```python
df = df2.applymap(remove_special)
```

```python
df.isnull().sum()
```

```python
df["Cuisines"].fillna("Unkown", inplace = True)
```

```python
df
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
categorical_columns = ["Has Table booking", "Has Online delivery", "Is delivering now", "Switch to order menu"]
df["Switch to order menu"].value_counts()
```

```python
df_encoded = pd.get_dummies(df, columns= categorical_columns, drop_first= True)
```

```python
df_encoded
```

```python
df_encoded.info()
```

```python
df_encoded["Rating text"].value_counts()
```

```python
label = {
    "Excellent" : 5,
    "Very Good" : 4,
    "Good" : 3,
    "Average" : 2,
    "Poor" : 1,
    "Not rated" : 0
}
df_encoded["Rating text"] = df["Rating text"].map(label)
```

```python
df_encoded
```

```python
df_encoded["Cuisines freq"] = df_encoded["Cuisines"].map(df["Cuisines"].value_counts(normalize= True))
```

```python
df_encoded
```

```python
label = LabelEncoder()
df_encoded["Rating color"] = label.fit_transform(df["Rating color"])
```

## 2. Feature Engineering

```python
columns = df_encoded.columns
columns
```

```python
x = df_encoded.drop(columns= ['Restaurant ID', 'Restaurant Name', 'Country Code', 'City', 'Address','Locality', 'Locality Verbose', 'Longitude', 'Latitude', 'Cuisines','Currency', 'Price range','Votes'])
scal = StandardScaler()
feat = ["Average Cost for two","Aggregate rating","Rating color","Rating text", "Cuisines freq"]
x[feat] = scal.fit_transform(x[feat])
x
```

```python
y = df_encoded["Cuisines"]
y
```

## 3. Splitting the dataset

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

## 4. Model fitting and training

Random Forest Classifier

```python
rf = RandomForestClassifier(n_estimators= 100)

rf.fit(x_train, y_train)
```

```python
y_pred_rf = rf.predict(x_test)
```

```python
print(f"{round(accuracy_score(y_pred_rf, y_test)*100, 2)}%")
```

Logistic Regression

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

```python
y_pred_log = model.predict(x_test)
print(f"{round(accuracy_score(y_pred_log, y_test)*100, 2)}%")
round(precision_score(y_test, y_pred_log, average= "macro")*100, 2)
```

Support Vector Machines


```python
svc = SVC()

svc.fit(x_train, y_train)
```

```python
y_pred_svm = svc.predict(x_test)
```

```python
print(f"The Accuracy for this model is:- {round(accuracy_score(y_pred_svm, y_test)*100, 2)}%")
print(f"The Precision for this model is:- {round(precision_score(y_test, y_pred_svm, average= "macro")*100, 2)}%")
print(f"The Recall for this model is:- {round(recall_score(y_pred_svm, y_test, average= "macro")*100, 2)}%")
```

Naive Bayes Model

```python
naive = GaussianNB()

naive.fit(x_train, y_train)
```

```python
y_pred_naive = naive.predict(x_test)
```

```python
print(f"The Accuracy for this model is:- {round(accuracy_score(y_pred_naive, y_test)*100, 2)}%")
print(f"The Precision for this model is:- {round(precision_score(y_test, y_pred_naive, average= "macro")*100, 2)}%")
print(f"The Recall for this model is:- {round(recall_score(y_pred_naive, y_test, average= "macro")*100, 2)}%")
```

```python
data = {
    "Model" : ["Random Forest", "Logistic Regression", "Support Vector Machines", "Naive Bayes"],
    "Accuracy" : [round(accuracy_score(y_pred_rf, y_test)*100, 2) , round(accuracy_score(y_pred_log, y_test)*100, 2), round(accuracy_score(y_pred_svm, y_test)*100, 2), round(accuracy_score(y_pred_naive, y_test)*100, 2)],
    "Precision" : [round(precision_score(y_test, y_pred_rf, average = "macro", zero_division=1)*100, 2) , round(precision_score(y_test, y_pred_log, average = "macro", zero_division=1)*100, 2) ,round(precision_score(y_test, y_pred_svm, average= "macro", zero_division=1)*100, 2) , round(precision_score(y_pred_naive, y_test, average= "macro", zero_division=1)*100, 2)],
    "Recall" : [round(recall_score(y_pred_rf, y_test, average = "macro", zero_division=1)*100, 2) , round(recall_score(y_pred_log, y_test, average = "macro", zero_division=1)*100, 2) , round(recall_score(y_pred_svm, y_test, average= "macro", zero_division=1)*100, 2), round(recall_score(y_pred_naive, y_test, average= "macro",zero_division=1)*100, 2)]
}

data_df = pd.DataFrame(data)

data_df
```

## 5. Hyperparameteric tuning of the models

Random Forest 

```python
param_grid_rf = {
    "n_estimators" : [100, 200 , 500],
    'max_depth' : [None, 10 ,20]
}

gridSearch = GridSearchCV(estimator= rf, param_grid= param_grid_rf, cv=5)
gridSearch.fit(x_train, y_train)

best_par = gridSearch.best_params_

best_est = gridSearch.best_estimator_
best_est.fit(x_train, y_train)

y_pred_rf_hyper = best_est.predict(x_test)
print(f"Best parameter: {best_par} and the best estimator : {best_est}")
print(f"The best estimator accuracy is:- {round(accuracy_score(y_pred_rf_hyper, y_test),2)*100}%")
```

Logistic Regression

```python
param_grid_log = {
    "penalty" : ["l1", "l2"],
    "C" : [0.1, 10, 100]
}

gridSearch_log = GridSearchCV(model, param_grid_log, cv=5)

gridSearch_log.fit(x_train, y_train)

best_par_log = gridSearch_log.best_params_

best_est_log = gridSearch_log.best_estimator_
best_est_log.fit(x_train, y_train)

y_pred_log_hyper = best_est_log.predict(x_test)
print(f"Best parameter: {best_par_log} and the best estimator : {best_est_log}")
print(f"The best estimator accuracy is:- {round(accuracy_score(y_pred_log_hyper, y_test),2)*100}%")
```

```python
param_grid_svm = {
    "kernel" : ["linear", "poly", "rbf"],
    "C" : [0.1,10,100]
}

gridSearch_svm = GridSearchCV(svc, param_grid_svm, cv=5)

gridSearch_svm.fit(x_train, y_train)

best_par_svm = gridSearch_svm.best_params_

best_est_svm = gridSearch_svm.best_estimator_
best_est_svm.fit(x_train, y_train)

y_pred_svm_hyper = best_est_svm.predict(x_test)
print(f"Best parameter: {best_par_svm} and the best estimator : {best_est_svm}")
print(f"The best estimator accuracy is:- {round(accuracy_score(y_pred_svm_hyper, y_test),2)*100}%")
```

```python
classification_reports = {}
unique_cuisines = set(y_test) 
for cuisine in unique_cuisines:
    index = y_test == cuisine
    y_correct_testset = y_test[index]
    y_pred_new = y_pred_naive[index]
    classification_reports[cuisine] = classification_report(y_correct_testset, y_pred_new, output_dict=True, zero_division=1)


for cuisine, report in classification_reports.items():
    print(f"Performance for {cuisine}:")
    print("Accuracy:", report['accuracy'])
    print("Precision:", report['weighted avg']['precision'])
    print("Recall:", report['weighted avg']['recall'])
    print("F1-score:", report['weighted avg']['f1-score'])
    print()
```

# There are multiple challenges for this task and some of them are:- 
* The dataset contains many features which are very densely populated leading to overfitting. 
* The dataset contains elements in a large majority and low minority manner as the majority elements overburden the minority elements leading for them to be super imposed
* The training and tuning of the data takes a long time due to the massive size created because of the encoding of the categorical columns
* Certain classes in the dataset are under represented causing biasness towards the majority classes.
